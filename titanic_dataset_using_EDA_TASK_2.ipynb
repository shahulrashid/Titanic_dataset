Histogram

```python
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(16, 12))

for i, col in enumerate(numeric_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(dataset[col].dropna(), kde=True, bins=30)
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.suptitle('Histograms of Numeric Features', fontsize=16, y=1.02)
plt.show()
```

BOXPLOT

```python
plt.figure(figsize=(16, 12))

for i, col in enumerate(numeric_cols):
    plt.subplot(4, 3, i + 1)
    sns.boxplot(x=dataset[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.suptitle('Boxplots of Numeric Features', fontsize=16, y=1.02)
plt.show()
```

```python
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = dataset[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

## **pairplot**


```python
selected_features = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

sns.pairplot(dataset[selected_features], hue='Survived', palette='Set2')
plt.suptitle('Pairplot of Selected Features Colored by Survival', y=1.02)
plt.show()
```

```python
print("\nSurvival Rate by Gender:")
print(dataset.groupby('Sex')['Survived'].mean())

sns.countplot(x='Sex', hue='Survived', data=dataset)
plt.title("Survival by Gender")
plt.show()
```

```python
print("\nSurvival Rate by Class:")
print(dataset.groupby('Pclass')['Survived'].mean())

sns.countplot(x='Pclass', hue='Survived', data=dataset)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=dataset)
plt.title("Fare vs Survival")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=dataset)
plt.title("Age vs Survival")
plt.show()
```

```python
dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch']

print("\nSurvival Rate by Family Size:")
print(dataset.groupby('FamilySize')['Survived'].mean())

sns.barplot(x='FamilySize', y='Survived', data=dataset)
plt.title("Survival Rate by Family Size")
plt.show()
```

```python
print("\nSurvival Rate by Embarked:")
print(dataset.groupby('Embarked')['Survived'].mean())

sns.countplot(x='Embarked', hue='Survived', data=dataset)
plt.title("Survival by Embarkation Port")
plt.show()
```

Questions & answer

1. What is the purpose of EDA (Exploratory Data Analysis)?
EDA is used to:

Understand the structure and patterns in the data.

Detect anomalies, missing values, and outliers.

Generate hypotheses and guide feature engineering.

Prepare data for modeling by identifying transformations needed.

👉 It’s the first step in any data science project to make data understandable and usable.

2. How do boxplots help in understanding a dataset?
Boxplots visualize:

Median, quartiles (Q1, Q3), and interquartile range (IQR).

Outliers (points outside 1.5×IQR).

Data spread and skewness.

👉 They’re great for comparing distributions across categories (e.g., age of survivors vs non-survivors).

3. What is correlation and why is it useful?
Correlation measures the strength and direction of a linear relationship between two variables:

Values range from -1 (strong negative) to +1 (strong positive).

0 means no linear relationship.

👉 It's useful to:

Detect relationships between features.

Identify redundant features (e.g., multicollinearity).

Select important variables for models.

4. How do you detect skewness in data?
Skewness indicates asymmetry in a distribution:

Right-skewed (positive): Tail on the right (e.g., income).

Left-skewed (negative): Tail on the left.

How to detect:

Visually using histograms or boxplots.

Statistically using .skew() in pandas or scipy.stats.skew().

5. What is multicollinearity?
Multicollinearity occurs when two or more independent variables are highly correlated.

Why it's a problem:

Inflates variances of coefficient estimates in regression.

Makes model interpretation difficult.

Detection tools:

Correlation matrix.

VIF (Variance Inflation Factor).

6. What tools do you use for EDA?
Common EDA tools in Python:

📊 Pandas – Data manipulation and summary.

📈 Matplotlib / Seaborn – Visualization.

📦 Plotly – Interactive plots.

📋 Missingno – Missing data visualization.

🔍 Scipy / Statsmodels – Statistical tests.

📊 Sweetviz / pandas-profiling / Autoviz – Automated EDA reports.

7. Can you explain a time when EDA helped you find a problem?
Example:
In a fraud detection project, EDA revealed:

An imbalance: only 0.17% transactions were fraud.

A key feature amount was skewed heavily.

One merchant ID was overrepresented with fraud.

This led to:

Using SMOTE for class balancing.

Transforming skewed features.

Flagging and down-weighting the fraudulent merchant ID in modeling.

8. What is the role of visualization in ML?
Visualizations help:

Understand feature distributions.

Detect class imbalance, outliers, skewness.

Reveal feature-target relationships.

Diagnose model errors (e.g., confusion matrix, residual plots).

Communicate insights effectively.

👉 Visualization is essential for interpretability, debugging, and decision-making.
